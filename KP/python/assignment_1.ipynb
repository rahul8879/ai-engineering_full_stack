```python
# Problem 4: AI Model Grader

score = float(input("Enter model accuracy score (0-100): "))

if score >= 90:
    print("Excellent! Production-ready.")
elif score >= 75:
    print("Good. Needs minor tuning.")
elif score >= 60:
    print("Fair. More training data needed.")
else:
    print("Poor. Re-evaluate your approach.")
```

```python
# Problem 5: Even or Odd Classifier

number = int(input("Enter a number: "))

# Even or Odd
if number % 2 == 0:
    print("Even")
else:
    print("Odd")

# Positive, Negative, or Zero
if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Problem 9: Sum of a List

numbers = [4, 17, 2, 9, 33, 7, 15]

total = 0
largest = numbers[0]

for n in numbers:
    print(n)
    total += n

    if n > largest:
        largest = n

print("Total sum:", total)
print("Largest number:", largest)
```

```python
# Problem 10: FizzBuzzAI

for i in range(1, 51):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# Problem 11: Number Guessing Game

import random

secret_number = random.randint(1, 20)
attempts = 0

while True:
    guess = int(input("Guess a number (1-20): "))
    attempts += 1

    if guess < secret_number:
        print("Too low!")
    elif guess > secret_number:
        print("Too high!")
    else:
        print("Correct!")
        print("Attempts:", attempts)
        break
```
