```python
# in interview

data = {
    'customer_id': ['BF001', 'BF002', 'BF003', 'BF004', 'BF005'],
    'age':         [28, 35, 42, 24, 31],
    'loan_amount': [250000, 500000, 150000, 750000, 320000],
    'emi':         [8500, 15200, 5100, 22000, 10500],
    'credit_score':[720, 680, 790, 610, 745],
    'city':        ['Mumbai', 'Pune', 'Delhi', 'Mumbai', 'Pune'],
    'status':      ['Approved', 'Approved', 'Rejected', 'Pending', 'Approved']
}

# create the dataframe
df = pd.DataFrame(data)


#city wise --> avg loan amount
df[df.city=='Mumbai']['loan_amount'].mean()
```

```python
def avg_valu_cal(df, column_name, city,avg_column_name):
    df1 = df[df[column_name]==city]
    result = df1[avg_column_name].mean()
    return result

avg_valu_cal(df,'city','Mumbai','loan_amount')
```

```python
output = {}
for i in df['city']:
    output[i] = avg_valu_cal(df,'city',i,'loan_amount')

output    
```
