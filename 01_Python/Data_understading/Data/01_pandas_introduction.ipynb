```python
def avg_valu_cal(df, column_name, city,avg_column_name):
    df1 = df[df[column_name]==city]
    result = df1[avg_column_name].mean()
    return result

avg_valu_cal(df,'city','Mumbai','loan_amount')
```

```python
output = {}
for i in df['city']:
    output[i] = avg_valu_cal(df,'city',i,'loan_amount')

output    
```
