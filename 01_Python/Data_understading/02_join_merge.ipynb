```python

customers = pd.DataFrame({
    'customer_id': [101, 102, 103, 104],
    'name':        ['Rahul', 'Priya', 'Arjun', 'Sneha'],
    'city':        ['Mumbai', 'Delhi', 'Pune', 'Bangalore'],
    'tier':        ['Gold', 'Silver', 'Gold', 'Bronze'],
})

orders = pd.DataFrame({
    'order_id':    [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    'customer_id': [101,  102,  101,  103,  102,  103,  101,  105 ],
    'product_id':  ['P01','P02','P03','P01','P02','P03','P01','P02'],
    'amount':      [1500, 3200, 800, 2100, 950, 1800, 2400, 600],
    'status':      ['delivered','pending','delivered','cancelled',
                    'delivered','delivered','pending','delivered'],
})

products = pd.DataFrame({
    'product_id':   ['P01', 'P02', 'P03'],
    'product_name': ['Laptop', 'Phone', 'Headphones'],
    'category':     ['Electronics', 'Electronics', 'Accessories'],
    'price':        [45000, 28000, 3500],
})
```

```python
# # total revenue per customer ??
df = pd.merge(orders,customers,on='customer_id',how='left')
print(df)
# df.head()
df.groupby('name').agg(
   { 'amount' : ['sum','mean']
    }
)


# # total revenue per cities ??

df.groupby('city').agg(
   { 'amount' : ['sum','mean']
    }
)
```

```python
# total order
# total spent
# avg order
# max order

result =df.groupby('name').agg(
    total_orders   = ('order_id', 'count'),
    total_spent    = ('amount',   'sum'),
    avg_order      = ('amount',   'mean'),
    max_order      = ('amount',   'max'),
)

result.reset_index().to_csv('final_result.csv')
```
