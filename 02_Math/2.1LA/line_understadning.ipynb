```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'income':[12,13,14,6,5,4],
    'credit_score':[750,780,800,300,400,460],
    'Status':['YES','YES','YES','NO','NO','NO']
}
df = pd.DataFrame(data)
df
```

```python
X = np.array(df.income)
X
```

```python
# plot it ??
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='income',y='credit_score',hue='Status')

w1=-0.45 # against income
w2=0.025 # aganst cibil socre
w0=-11



income_vals = np.array([X.min() - 5, X.max() + 5])
score_vals = (-w1 * income_vals - w0) / w2


plt.plot(income_vals,score_vals)
plt.grid()
plt.show()
```
