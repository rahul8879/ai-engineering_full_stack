```python
customer_info = [[17,650,13],[16,560],[18,780]]
new_customer = [18,760,13]
import math
# compute the distance between each points
distance = []
for i in range(len(customer_info)):
    value = math.sqrt((customer_info[i][0]-new_customer[0])**2 + (customer_info[i][1]-new_customer[1])**2)
    distance.append(value)
```
