```python
# --- A1 ---
a1 = pd.merge(orders, customers, on='customer_id', how='left')
print('A1:', a1[['order_id','customer_id','name','city','amount','status']].to_string(index=False))
```

```python
# --- A2 ---
a2 = pd.merge(customers, orders, on='customer_id', how='left')
no_orders = a2[a2['order_id'].isna()][['customer_id','name','city']]
print('A2 - customers with no orders:')
print(no_orders.to_string(index=False))
```

```python
# --- A3 ---
enriched = (
    orders
    .pipe(pd.merge, customers, on='customer_id', how='left')
    .pipe(pd.merge, products,  on='product_id',  how='left')
)
print('A3:', enriched[['order_id','name','city','tier','product_name','category','amount','status']]
      .to_string(index=False))
```

```python
# --- A4 ---
enriched['profit_margin']    = ((enriched['sell_price'] - enriched['cost_price']) / enriched['sell_price'] * 100).round(1)
enriched['revenue_vs_list']  = (enriched['amount'] / enriched['sell_price'] * 100).round(1)
print('A4:', enriched[['order_id','name','product_name','sell_price','amount','profit_margin','revenue_vs_list']]
      .to_string(index=False))
```

```python
# --- A5 ---
print('Revenue per city:')
print(enriched.groupby('city')['amount'].sum().sort_values(ascending=False).reset_index())
print('\nOrders per status:')
print(enriched.groupby('status')['order_id'].count().reset_index())
print('\nAvg order per tier:')
print(enriched.groupby('tier')['amount'].mean().round(0).reset_index())
```

```python
# --- A6 ---
scorecard = (
    enriched.dropna(subset=['name'])
    .groupby('name')
    .agg(
        total_orders      = ('order_id', 'count'),
        total_spent       = ('amount',   'sum'),
        avg_order         = ('amount',   lambda x: round(x.mean(), 0)),
        max_single_order  = ('amount',   'max'),
        delivered_count   = ('status',   lambda x: (x == 'delivered').sum()),
    )
)
print('A6:', scorecard)
```

```python
# --- A7 ---
enriched['customer_total']         = enriched.groupby('name')['amount'].transform('sum')
enriched['customer_order_rank']    = enriched.groupby('name')['amount'].rank(ascending=False).astype(int)
enriched['pct_of_customer_total']  = (enriched['amount'] / enriched['customer_total'] * 100).round(1)
enriched['is_top_order']           = enriched['customer_order_rank'] == 1
print('A7:', enriched[['order_id','name','amount','customer_total',
                        'customer_order_rank','pct_of_customer_total','is_top_order']]
      .to_string(index=False))
```

```python
# --- A8 ---
def city_report(g):
    return pd.Series({
        'num_customers':     g['name'].nunique(),
        'total_revenue':     g['amount'].sum(),
        'top_product':       g.groupby('product_name')['amount'].sum().idxmax() if g['product_name'].notna().any() else None,
        'cancellation_rate': round((g['status'] == 'cancelled').mean() * 100, 1),
        'avg_order_value':   round(g['amount'].mean(), 0),
    })
print('A8:', enriched.groupby('city').apply(city_report))
```

```python
# --- A9 ---
high_spenders = enriched.groupby('name').filter(lambda g: g['amount'].sum() > 3000)
print('A9a - High spenders:')
print(high_spenders[['order_id','name','amount']].to_string(index=False))

multi_order_cities = enriched.groupby('city').filter(lambda g: len(g) > 1)
print('\nA9b - Cities with >1 order:')
print(multi_order_cities[['order_id','city','amount']].to_string(index=False))
```

```python
# --- A10 Capstone ---
# Start from products (RIGHT table) to keep P04 with no orders
full = pd.merge(enriched, products[['product_id','product_name']].drop_duplicates(),
                on=['product_id','product_name'], how='right')

def product_report(g):
    return pd.Series({
        'total_units_sold':      g['order_id'].count(),
        'total_revenue':         g['amount'].sum(),
        'avg_revenue_per_order': round(g['amount'].mean(), 0) if g['amount'].count() > 0 else 0,
        'margin_pct':            round((g['sell_price'].iloc[0] - g['cost_price'].iloc[0]) / g['sell_price'].iloc[0] * 100, 1) if g['sell_price'].notna().any() else None,
        'top_customer':          g.groupby('name')['amount'].sum().idxmax() if g['name'].notna().any() else None,
        'dominant_city':         g.groupby('city')['order_id'].count().idxmax() if g['city'].notna().any() else None,
    })

prod_full = pd.merge(full, products[['product_id','cost_price','sell_price']], on='product_id', how='left', suffixes=('','_ref'))
print('A10 Capstone:')
print(enriched.groupby('product_name').apply(product_report))
```
